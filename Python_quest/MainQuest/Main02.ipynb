```python
#  Q1. 코어 명단(딕셔너리)을 입력받아 학번을 부여하는 메서드를 작성해보세요(이름: 학번, 이름:학번, …)
# {김건: 0, 김경민: 0, 김소연: 0, 김영신: 0, 박근수: 0, 박영수: 0, 박영준: 0, 서원영: 0, 신지만: 0, 심지안: 0, 이규상: 0, 이윤상: 0, 이진영: 0, 임현석: 0, 전요한: 0, 전재영: 0, 조대희:0}
# Q2. Aiffel 클래스로부터 생성된 학번의 개수를 출력하는 get_id_num() 메서드를 추가하세요.
# Q3. Aiffel 클래스에 퀘스트 점수 입력을 위한 score 메서드를 추가하세요. 점수는 +1점 부터 +3점까지만 가능합니다. 추가 점수를 관리하는 변수도 만들어주세요.
# Q4. Aiffel 클래스에 퀘스트 점수 차감을 위한 penalty 메서드를 추가하세요. penalty로 인한 잔여 점수는 음수가 가능합니다. 차감된 점수를 관리하는 변수도 만들어주세요.
# Q5. Aiffel 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 해당 메서드를 호출하면 이름, 학번, 과정종류, 퀘스트 점수 총점, 퀘스트 입력 총점, 퀘스트 penalty 총합이 새로운 코어 명단(csv파일)로 출력되어야 합니다.
# Q6. 중도포기자를 관리하기 위해 그루 정보를 삭제하는 메서드를 추가해보세요. 해당 메서드를 호출하면 명단이 갱신되어야 합니다.
# 김경민, 김영신 님을 삭제한후 갱신된 명단을 출력하세요.

# Q7. Aiffel 클래스를 상속하여, 그루 명단을 바탕으로 2인, 3인, 4인으로 랜덤 그룹핑을 해주는 메서드가 추가된 Aiffel_Group 클래스를 만들어보세요.
#  그룹 이름은 총 8개 입니다. (상생, 신뢰, 열정, 이끔, 성장, 가치, 공유, 확산)
# Q8. Aiffel 클래스를 상속하여, 그루 명단을 바탕으로 “힌튼”, “알트만”, “응”, “하사비스” 네개의 길드로 그루들을 랜덤그룹핑해주는 Aiffel_Guild 클래스를 만들어보세요.
# Q9. Aiffel_Guild 클래스에 길드별 점수를 입력하고 차감하는 guild_score 메서드를 추가해보세요. guild_score 메서드는 길드 이름과 점수를 인자로 받고 길드별 스코어를 출력합니다. guild_score 메서드로 점수가 입력되거나 차감되면, Aiffel 클래스에서 구현된 개별 그루의 score도 소속 길드에서 추가되거나 차감된 점수에 따라 별도로 증감되어야 합니다.
# ex. 모든 길드의 현재 점수가 0점이고 박영수, 박근수, 심지안 그루의 개별 점수가 각각 2, 3, 1 이며 박영수, 심지안 그루는 힌튼 길드, 박근수 그루는 응 길드라고 했을 때, 힌튼길드가 -1점, 응길드가 +1점을 받았다면 길드 점수는 힌튼 -1, 응 +1, 하사비스 0, 알트만 0 이 되어야하고 박영수, 심지안 그루는 각각 1, 0점 박근수 그루는 4점이 되어야 합니다.

# Q10. Aiffel_Group 클래스에 group2guild_score 메서드를 추가하세요. group2guild_score 메서드는 아래와 같이 작동해야 합니다.
# group2guild_score 메서드는 그룹이름과 그룹점수 2개의 인자를 받습니다.
# 랜덤으로 그룹핑된 그루들이 소속된 그룹의 이름으로 점수를 받게 되면, 각자 소속된 길드의 점수로 연동되어 계산되게 합니다.
# ex. 박영수, 박근수, 심지안 세명이 “상생” 그룹으로 그룹핑 되었을 때,
# 상생 그룹이 +2점을 받게 되면
# 힌튼 길드와 응길드의 각 길드 스코어는 +2점이 됩니다
# Q11. Aiffel_Group 클래스와 Aiffel_Guild 클래스가 Aiffel클래스의 display_info() 메서드를 아래와 같이 오버라이딩하도록 합니다.
# guild_score(), group2guild_score() 메서드를 호출했을 시
# 기존 명단에 길드별 스코어 총합이 보이도록 명단을 출력하도록 display_info() 를 오버라이딩합니다.

# 아이펠 캠퍼스에서는 그루들에게 학번을 부여하고 퀘스트 점수를 관리하고 있습니다. 이제부터 학번과 점수를 관리할 수 있는 Aiffel 클래스를 만들어보도록 하겠습니다.
# 이니셜라이저에서는 아래와 같은 4개의 인자를 받습니다.
# 이름, 입학연도, 기수, 과정명
# 학번은 7자리로 이뤄져있고 학번 생성 규칙은 아래와 같습니다
# YYNNCXX
# YY : 연도 마지막 두 자리 (ex. 2023년은 23) NN : 기수 (25기) C : 과정 종류 (리서치=1, 코어=2) XX : 가나다순 번호

```

```python
# 이름, 입학연도, 기수, 과정명
class Aiffel:
    def __init__(self, name, yr, term, course):
        self.name = name
        self.yr = yr
        self.term = term
        self.course = course
        self.student_id = self.generate_student_id()

# 학번 생성 규칙
    def generate_student_id(self):
        yr_last_two = str(self.yr)[-2:]    # 연도 마지막 두자리를 인덱싱
        term_number = str(self.term).zfill(2) # NN부분: 기수가 한 자리수이면 앞에 0을 추가함으로 써 '05'식으로 표현
        
        # 과정 종류(C)
        if self.course == "리서치":
            course_number = "1"
        elif self.course == "코어":
            course_number = "2"

        # 이름의 맨 앞글자의 아스키코드 값을 26으로 나눈 나머지에 1을 더하여 A~Z까지 즉 1~26으로 숫자 매김.
        # {:02d} : 두자리수로 표현. 예) 5 -> 05
        # 하지만 YYNNC가 동일하고 성이 같은 이름이 둘 이상 있을 때, 학번이 같아질 우려가 있음...
        student_id_suffix = "{:02d}".format(ord(self.name[0]) % 26 + 1)
        
        return yr_last_two + term_number + course_number + student_id_suffix

# 학생 리스트
A = Aiffel("홍길동", 2023, 25, "리서치")
print(A.student_id)  
B = Aiffel("홍철수", 2024, 26, "코어")
print(B.student_id) 
C = Aiffel("홍철민", 2024, 26, "코어")
print(C.student_id) 

```

```python
#  Q1. 코어 명단(딕셔너리)을 입력받아 학번을 부여하는 메서드를 작성해보세요(이름: 학번, 이름:학번, …)

class Aiffel:
    def __init__(self, name, yr, term, course):
        self.name = name
        self.yr = yr
        self.term = term
        self.course = course
        self.student_id = None

    def generate_student_id(self):
        year_last_two_digits = str(self.yr)[-2:]
        term_number = str(self.term).zfill(2)
        
        if self.course == "리서치":
            course_number = "1"
        elif self.course == "코어":
            course_number = "2"
            
        if len(self.name) >= 2:
            student_id_suffix = "{:02d}".format(ord(self.name[1]) % 26 + 1)
        else:
            student_id_suffix = "00"  # 이름이 한 글자일 경우 기본값으로 설정
        
        return year_last_two_digits + term_number + course_number + student_id_suffix

# 코어 학생 명단
A = Aiffel("김상길", 2023, 25, "코어")
B = Aiffel("홍철수", 2024, 26, "코어")
C = Aiffel("홍두민", 2024, 26, "코어")

Core_member = {
    A.name: {
    "admission_year": A.yr,
    "term": A.term,
    "course": A.course
    },
    B.name: {
    "admission_year": B.yr,
    "term": B.term,
    "course": B.course
    },
    C.name: {
    "admission_year": C.yr,
    "term": C.term,
    "course": C.course
    }
}
for name in Core_member.keys():
    member_info= Core_member[name]
    
    XD= Aiffel(name, member_info["admission_year"], member_info["term"], member_info["course"])
    XD.student_id= XD.generate_student_id()
    
    print(f"{name}: {XD.student_id}")
```

```python
# Q2. Aiffel 클래스로부터 생성된 학번의 개수를 출력하는 get_id_num() 메서드를 추가하세요.
# Q3. Aiffel 클래스에 퀘스트 점수 입력을 위한 additional_score 메서드를 추가하세요. 점수는 +1점 부터 +3점까지만 가능합니다. 추가 점수를 관리하는 변수도 만들어주세요.
# Q4. Aiffel 클래스에 퀘스트 점수 차감을 위한 penalty 메서드를 추가하세요. penalty로 인한 잔여 점수는 음수가 가능합니다. 차감된 점수를 관리하는 변수도 만들어주세요.


class Aiffel:
    ID_num = 0

    def __init__(self, name, yr, term, course):
        self.name = name
        self.yr = yr
        self.term = term
        self.course = course
        self.student_id = None
        self.additional_score = 0     # 추가 점수 초기화
        self.deducted_score = 0      # 차감 점수 초기화
        self.score = 0              # 점수 초기화

        Aiffel.ID_num += 1

    def generate_student_id(self):
        year_last_two_digits = str(self.yr)[-2:]
        term_number = str(self.term).zfill(2)
        
        if self.course == "리서치":
            course_number = "1"
        elif self.course == "코어":
            course_number = "2"
            
        if len(self.name) >= 2:
            student_id_suffix = "{:02d}".format(ord(self.name[1]) % 26 + 1)
        else:
            student_id_suffix = "00"  # 이름이 한 글자일 경우 기본값으로 설정
        
        return year_last_two_digits + term_number + course_number + student_id_suffix
    
    @classmethod
    def get_id_num(cls):  # 학번 개수 출력
        print(cls.ID_num)
    
    def score_up(self, up):  # 점수 입력
        if 1 <= up <= 3:
            self.additional_score += up
            self.score = self.additional_score - self.deducted_score
        else:
            print("점수는 한 회당 1점 이상, 3점 이하까지 부여됩니다")  # 1~3점이 아닌 점수를 주게 되는 경우
    
    def penalty(self, down): # 점수 차감
        if down > 0 :
            self.deducted_score += down
            self.score = self.additional_score - self.deducted_score

# 학생 명단
A = Aiffel("김상길", 2023, 25, "리서치")
B = Aiffel("홍철수", 2024, 26, "코어")
C = Aiffel("홍두민", 2024, 26, "코어")

A.get_id_num() # 총 학생 명단(학번) 수
print("")
B.score_up(2)
B.score_up(5)
B.score_up(1)
print("")
B.penalty(4)

print(f"총 추가점수: {B.additional_score}점")
print(f"총 차감된 점수: {B.deducted_score}점")
print(f"점수 합계: {B.score}점")
```
