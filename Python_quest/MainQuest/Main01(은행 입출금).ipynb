```python
# MainQuest Q10. 입출금 내역 (최종!!)

import random

class Account:
    bank = "SC은행" #은행이름
    num_accounts = 0

    #예금주, 계좌번호, 잔액 설정(Q1)
    def __init__(self, owner, initial_balance):
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []
        
        self.owner = owner
        self.account_number = self.generate_account_number()
        self.balance = initial_balance

        Account.num_accounts += 1 # 생성된 계좌 객체 개수 저장(Q2)

    # 계좌번호는 랜덤 생성
    @staticmethod
    def generate_account_number():
        account_number = f"{random.randint(0, 999):03d}-{random.randint(0, 99):02d}-{random.randint(0, 999999):06d}" # 3자리-2자리-6자리
        return account_number
    
    @classmethod 
    def get_account_num(cls): # 변수 출력 메서드(Q3)
        print(cls.num_accounts)
    
    def deposit(self, amount): # 입금 메서드 추가(Q4)
        if amount > 0:
            self.deposit_log.append(amount)
            self.balance += amount
            
            self.deposit_count += 1
            if self.deposit_count % 5 == 0:    # 5회마다 입금시 이자 지급(Q7)
                self.balance = (self.balance * 1.01)

    def withdraw(self, amount): # 출금 메서드 추가(Q5)
        if amount > 0 and amount <= self.balance:
            self.withdraw_log.append(amount)
            self.balance -= amount
            return self.balance
        else: 
            return "잔액이 부족합니다." # 출금이 잔액보다 많을 경우

    def display_info(self): #정보 출력 메서드 추가(Q6)
        print("은행이름: ", self.bank)
        print("예금주: ", self.owner)
        print("계좌번호: ", self.account_number)
        print("잔고: ", f"{self.balance:,}")
    
    def withdraw_history(self):
        for amount in self.withdraw_log:
            print(amount)

    def deposit_history(self):
        for amount in self.deposit_log:
            print(amount)

# Q8. 여러 객체 생성 Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.
account_list=[]
HAS = Account("하사비스",10000)
ALT = Account("알트만",20000)
HIN = Account("힌튼",30000)
JDH = Account("지동현",40000)

account_list.append(HAS)
account_list.append(ALT)
account_list.append(HIN)
account_list.append(JDH)

JDH.deposit(20000)
JDH.deposit(30000)
JDH.deposit(40000)
JDH.deposit_history()

print(f"잔액: {JDH.balance}원")

JDH.withdraw(15000)
JDH.withdraw(12000)
JDH.withdraw(37000)
JDH.withdraw_history()

print(f"잔액: {JDH.balance}원")
```
